```python
import os

dataset_path = dataset.location

def count_images(subset):
    path = os.path.join(dataset_path, subset, 'images')
    if os.path.exists(path):
        return len([file for file in os.listdir(path) if file.endswith(('.jpg', '.jpeg', '.png'))])
    return 0

train_count = count_images('train')
valid_count = count_images('valid')
test_count = count_images('test')

print(f"Nombre d'images dans le dataset :")
print(f"Train : {train_count}")
print(f"Validation : {valid_count}")
print(f"Test : {test_count}")
```

```python
import os
from collections import defaultdict

dataset_path = dataset.location

class_counts = defaultdict(int)

def count_classes(subset):
    path = os.path.join(dataset_path, subset, 'labels')
    if os.path.exists(path):
        for file in os.listdir(path):
            if file.endswith('.txt'):
                with open(os.path.join(path, file), 'r') as f:
                    for line in f:
                        class_id = line.split()[0]  # La première colonne est l'ID de la classe
                        class_counts[class_id] += 1

count_classes('train')
count_classes('valid')
count_classes('test')

print("Nombre d'instances par classe :")
for class_id, count in sorted(class_counts.items()):
    print(f"Classe {class_id} : {count} instances")
```
